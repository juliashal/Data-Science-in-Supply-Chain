8- make a subset of the data that has the car name, the price and name the new sub-setted data frame  car pricing. <br>
9- create a function called pricing category that returns "Budget Car " if the cars are less than 20,000 USD," Suitable Car " is the car is more than 20,000 and less than 35 000 and finally an expensive car for cars more than 35000.  <br>
10- create a column named category on the subset using a for loop and pricing category function. <br>
11- How many Budget cars, suitable cars, and expensive cars we have? <br>


```python
df_short = df[['car name','Price']]
```

```python
def price_category(price):
    if price < 20_000:
        return "Budget"
    elif price < 35_000 and price >= 20_000:
        return "Suitable"
    else:
        return "Expensive"
```

```python
df_short['Category'] = df_short['Price'].map(price_category)
```
